```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from pathlib import Path


# 1. Load the cleaned dataset

CLEAN_PATH = Path(r"C:\Users\pooja\OneDrive\Desktop\PoojaPattu_E_DA_asses\Data\cleaned and preprocessed data\UberDataset_Cleaned.csv")
PROCESSED_PATH = Path(r"C:\Users\pooja\OneDrive\Desktop\PoojaPattu_E_DA_asses\Data\cleaned and preprocessed data\UberDataset_Preprocessed.csv")

print("üöó Loading cleaned dataset...")
df = pd.read_csv(CLEAN_PATH)
print(f"Initial shape: {df.shape}")
print("=" * 60)
```

```python
# 2. Encode categorical variables

from sklearn.preprocessing import LabelEncoder
label_cols = ['CATEGORY', 'START', 'STOP', 'PURPOSE']
encoder = LabelEncoder()

for col in label_cols:
    df[col] = df[col].astype(str).str.strip()
    df[col + '_ENC'] = encoder.fit_transform(df[col])
    print(f"‚úÖ Encoded: {col} ‚Üí {col}_ENC")
```

```python
# 4. Scale numeric features

scaler = MinMaxScaler()
num_cols = ['MILES', 'TRIP_DURATION_MIN']

df[num_cols] = scaler.fit_transform(df[num_cols])
print(f"‚úÖ Scaled numerical columns: {num_cols}")
```

```python
# 5. Validation checks

assert df.isna().sum().sum() == 0, "‚ùå Missing values remain after preprocessing!"
print("‚úÖ Validation passed ‚Äî no missing values found.")
print(f"Final dataset shape: {df.shape}")
```

```python
# 6. Save preprocessed dataset

PROCESSED_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(PROCESSED_PATH, index=False)
print(f"‚úÖ Preprocessed dataset saved successfully at:\n{PROCESSED_PATH}")
```
